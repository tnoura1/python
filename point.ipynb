```python
import numpy as np

def calculate_grades():
    num_students = int(input("Enter the number of students: "))
    num_subjects = int(input("Enter the number of subjects: "))
    marks = np.zeros((num_students, num_subjects))
    for i in range(num_students):
        print(f"\nEnter marks for student {i+1}:")
        for j in range(num_subjects):
            marks[i, j] = float(input(f"Enter marks for subject {j+1}: "))
    total_marks = np.sum(marks, axis=1)
    percentages = (total_marks / (num_subjects * 100)) * 100
    grades = np.where(percentages >= 90, 'A+',
                      np.where(percentages >= 80, 'A',
                               np.where(percentages >= 70, 'B+',
                                        np.where(percentages >= 60, 'B',
                                                 np.where(percentages >= 50, 'C', 'F')))))
    print("\nResults:")
    print("---------")
    print("Student\tTotal Marks\tPercentage\tGrade")
    for i in range(num_students):
        print(f"{i+1}\t\t{total_marks[i]}\t\t{percentages[i]:.2f}%\t\t{grades[i]}")
calculate_grades()
```
